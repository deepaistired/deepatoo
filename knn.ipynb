```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data=[2,3,4,10,11,12,20,25,30], columns=["data"])
print(df)
```

```python
print("Assuming the initial means m1=4 and m2=14")
print("Calculating new cluster points")
df["Dist m1"]= abs(4-df["data"])
df["Dist m2"]= abs(12-df["data"])
df["Cluster"] = np.where(df["Dist m1"] < df["Dist m2"], "k1", "k2")
df
```

```python
means = df.groupby('Cluster')['data'].mean()
print("New means:", means)
```
